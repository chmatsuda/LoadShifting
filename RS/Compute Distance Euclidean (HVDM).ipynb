# k=1

```python
RecSched = np.zeros(15)
for testbuilding in range (15):
    TB = TBs[testbuilding]
    Distances=[0,0,0,0,0]
    for j in range(5):
        X = [RBs[j],TB]
        distance= (dist.pairwise(X))[0][1]
        #print(distance)
        Distances[j] = distance
    print(Distances)
    minpos = Distances.index(min(Distances))
    RecSched[testbuilding]=minpos+1
    print("Recommend Sch " + str(minpos+1)+" for Test Building "+str(testbuilding+1))
```

# K=2

```python
RecSched2 = np.zeros(15)
for testbuilding in range (15):
    TB = TBs[testbuilding]
    Distances=[0,0,0,0,0]
    DistancesCopy=[0,0,0,0,0]
    for j in range(5):
        X = [RBs[j],TB]
        distance= (dist.pairwise(X))[0][1]
        #print(distance)
        Distances[j] = distance
        DistancesCopy[j]=distance
    print(Distances)
    min1 = min(DistancesCopy)
    #print("min1 "+ str(min1))
    minpos = DistancesCopy.index(min1)
    DistancesCopy[minpos] = min1+2
    #print(DistancesCopy)
    min2 = min(DistancesCopy)
    #print("min2 "+ str(min2))
    minpos2 = DistancesCopy.index(min2)
    RecSched[testbuilding]=minpos+1
    RecSched2[testbuilding]=minpos2+1
    #print(minpos2+1)
    print("Recommend Sch " + str(minpos+1)+" & "+ str(minpos2+1)+" for Test Building "+str(testbuilding+1))
```
