```python
import os
import opyplus as op
import numpy as np

eplus_dir_path = op.get_eplus_base_dir_path((9, 6, 0))
```

# # Running EP Simulation

```python
idf_path = os.path.join(
    eplus_dir_path,
    "ExampleFiles",
    "HPSim",
    "5_US+SF+CZ5B+hp+crawlspace+IECC_2021.expidf")
epw_path = os.path.join("-w",
    eplus_dir_path,
    "WeatherData",
    "USA_CO_Denver-Aurora-Buckley.AFB.724695_TMY3.epw")

# # run simulation
s = op.simulate(idf_path, epw_path, "tester")
```

```python
print(f"status: {s.get_status()}\n")
print(f"Eplus .err file:\n{s.get_out_err().get_content()}")
```

```python
# retrieve hourly output (.eso file)
hourly_output = s.get_out_eso()

# ask for datetime index on year 2013
hourly_output.create_datetime_index(2021)

# get Pandas dataframe
df = hourly_output.get_data()
print(hourly_output.get_info())
```

```python
##Zone Temperature ###
ZoneT = df[["living_unit1,Zone Mean Air Temperature"]]
OutsideT = df["environment,Site Outdoor Air Drybulb Temperature"]
Heating = df["heating:electricity,Meter"]


Notmet = sum (df["living_unit1,Zone Heating Setpoint Not Met While Occupied Time"])
# monthly resample and display
# print(df[[
#     "living_unit1,Zone Mean Air Temperature",
# ]])
print(Notmet)
```
